```python
import re
import string

def capitalize_each_word(original_str):
    result = ""
    # Split the string and get all words in a list
    list_of_words = original_str.split()
    # Iterate over all elements in list
    for elem in list_of_words:
        # capitalize first letter of each word and add to a string
        if len(result) > 0:
            result = result + " " + elem.strip().capitalize()
        else:
            result = elem.capitalize()
    # If result is still empty then return original string else returned capitalized.
    if not result:
        return original_str
    else:
        return result

def main():
    # read file line by line
    file = open( "/content/input.srt", "r")
    lines = file.readlines()
    file.close()

    text = ''
    for line in lines:
        if re.search('^[0-9]+$', line) is None and re.search('^[0-9]{2}:[0-9]{2}:[0-9]{2}', line) is None and re.search('^$', line) is None:
              for line in line.split("\n"):
                print(line)

                if line != "":
                  if not "[ __ ]" in line and "[Music]":
                    text += line.rstrip("\n")[:1].title()
                    text += str(line.rstrip("\n")[1:])
                    if text.endswith(".") == False:
                      text += "."
                else:
                    pass
                text += "\n"


                #text += line.rstrip("\n")[:1].title() + str(line.rstrip("\n")[1:]) + "\n"
        text = text.lstrip()
    print(text)

main()
```

```python
# @markdown # Initialize functions
%cd /content/
!pip install pydub~=0.25.1 --upgrade
!pip install SpeechRecognition --upgrade
import speech_recognition as sr
import os
from pydub import AudioSegment
from pydub.silence import split_on_silence
import re
import sys
from pydub import AudioSegment
from pydub.silence import split_on_silence
import glob
import os
import wave
import contextlib
import time

# create a speech recognition object
r = sr.Recognizer()

#@markdown experiment with this value for your target audio file
minimum_silence_length=100 #@param {type: "number"}

#@markdown How quiet each end of sentence is (-DB)
silence_threshold =  31#@param {type: "number"}

#@markdown keep the silence for 1 second, adjustable as well
keep_silence_len =  250#@param {type: "number"}

!mkdir input/

# a function that splits the audio file into chunks
# and applies speech recognition
def get_large_audio_transcription(path):
    """
    Splitting the large audio file into chunks
    and apply speech recognition on each of these chunks
    """
    # open the audio file using pydub
    sound = AudioSegment.from_wav(path)
    # split audio sound where silence is 700 miliseconds or more and get chunks
    chunks = split_on_silence(sound,
                              min_silence_len= minimum_silence_length,
                              silence_thresh=sound.dBFS - silence_threshold, 
                              keep_silence= keep_silence_len, 
                              )
    folder_name = "audio-chunks"
    # create a directory to store the audio chunks
    if not os.path.isdir(folder_name):
        os.mkdir(folder_name)
    whole_text = ""
    # process each chunk 
    for i, audio_chunk in enumerate(chunks, start=1):
        # export audio chunk and save it in
        # the `folder_name` directory.
        chunk_filename = os.path.join(folder_name, f"{i}.wav")
        audio_chunk.export(chunk_filename, format="wav")
        # recognize the chunk
        with sr.AudioFile(chunk_filename) as source:
            audio_listened = r.record(source)
            # try converting it to text
            try:
                text = r.recognize_google(audio_listened)
            except sr.UnknownValueError as e:
                print("Error:", str(e))
            else:
                text = f"{text.capitalize()}. "
                print(chunk_filename, ":", text)
                whole_text += text
    # return the text for all chunks detected
    return whole_text
```

```python
#@title Put your audio file into input/ then enter the audio name (include filetype but not path)
#@markdown Results will be saved in output/

audio_file = 'Pain.wav'  #@param {type: "string"}

#@markdown #### Set to true for more accurate results, set to false if your audio file is too small.

split_audio_into_chunks = "True" #@param ["True", "False"]

skip_large_duration_files = "False" #@param ["True", "False"]

#@markdown ---

file_number = 1

# assign files
input_file = 'input/' + audio_file

# create dir if doesn't exist
os.makedirs(os.path.dirname('input/'), exist_ok=True)
os.makedirs(os.path.dirname('output/'), exist_ok=True)
os.makedirs(os.path.dirname('output/wavs/'), exist_ok=True)

print("Splitting audio into chunks, may take a while.")

sound_file = AudioSegment.from_file(input_file)
sound_file = sound_file.set_frame_rate(22050)  # don't change this
sound_file = sound_file.set_channels(1)  # don't change this

if split_audio_into_chunks is True:
  audio_chunks = split_on_silence(sound_file, min_silence_len=500,  # 1000 cuts at 1 second of silence. 500 is 0.5 sec
                                silence_thresh=-40)
else:
  audio_chunks = split_on_silence(sound_file, min_silence_len=2000,  # 1000 cuts at 1 second of silence. 500 is 0.5 sec
                                silence_thresh=-40)

print("Beggining transcription!")

for i, chunk in enumerate(audio_chunks):

    if len(os.listdir('output/wavs')) == 0:
        print("Wavs directory is empty")
    else:
        list_of_files = glob.glob('output/wavs/*')  # * means all
        latest_file = max(list_of_files, key=os.path.getctime)

        # Extract numbers and cast them to int
        list_of_nums = re.findall('\\d+', latest_file)

        if int(list_of_nums[0]) >= file_number:
            file_number = int(list_of_nums[0]) + 1

    out_file = "output/wavs/{0}.wav".format(file_number)
    print("exporting", out_file)

    chunk.export(out_file, format="wav")

    fname = out_file
    with contextlib.closing(wave.open(fname, 'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        print('Duration:', duration)

    if skip_large_duration_files:
        if duration < 12:
            start_time = time.time()
            transcription = get_large_audio_transcription(out_file)

            if transcription != '':
                if os.path.isfile('output/list.txt'):
                    if os.stat("output/list.txt").st_size != 0:
                        with open('output/list.txt', 'a+') as f:
                            f.write(f'\nwavs/{file_number}.wav|' + transcription)
                            f.flush()
                    else:
                        with open('output/list.txt', 'a+') as f:
                            f.write(f'wavs/{file_number}.wav|' + transcription)
                            f.flush()
                else:
                    with open('output/list.txt', 'x') as f:
                        f.write(f'wavs/{file_number}.wav|' + transcription)

                file_number = file_number + 1
            else:
                os.remove(out_file)
        else:
            os.remove(out_file)

    else:
        start_time = time.time()
        transcription = get_large_audio_transcription(out_file)

        if transcription != '':
            if os.path.isfile('output/list.txt'):
                if os.stat("output/list.txt").st_size != 0:
                    with open('output/list.txt', 'a+') as f:
                        f.write(f'\nwavs/{file_number}.wav|' + transcription)
                        f.flush()
                else:
                    with open('output/list.txt', 'a+') as f:
                        f.write(f'wavs/{file_number}.wav|' + transcription)
                        f.flush()
            else:
                with open('output/list.txt', 'x') as f:
                    f.write(f'wavs/{file_number}.wav|' + transcription)

            file_number = file_number + 1
        else:
            os.remove(out_file)

end_time = time.time()

time_lapsed = end_time - start_time

print("\n=============\nFinished transcribing!\n")
print("Took (seconds): "+str(time_lapsed)+ "\n")
print("=============\n")
```
